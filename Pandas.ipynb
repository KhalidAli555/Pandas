# Working with Missing Data

```python
import pandas as pd
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
data=pd.read_csv(r"company1.csv")
```

```python
data
```

```python
print(data.isnull())    # null value return true
```

```python
print(data.isnull().sum())   # return how many column have null values
```

```python
print(data.dropna())   # dropna remove the whole row who have NA's(but this method is good when dataset is large otherwise it is not correct)
```

```python
print(data.fillna(130))    # fill 130 in all columns who have NA's
```

```python
print(data["salary"].fillna(30000))  # Add 300000 value in 'salary' column
```

```python
data
```

```python
x=data['salary'].mean()     #replace null values by mean ,(inplace=True make changes in original dataset)
data['salary'].fillna(x)
```

```python
data
```

```python
print(data.fillna(method="bfill"))    # fill backward value in dataset(means bottom to top)
```

```python
print(data.fillna(method="ffill"))   # fill forward value in dataset(means top to bottom)
```

# Merge,Concatenate & Join DataFrame

```python
data1={"Emp Id":["101","102","103","104","105","106"],
       "Name":["khlaid","jameel","Umar","Javid","Haseeb","ashar"],
       "Age":[20,30,40,50,60,45]}
```

```python
data2={"Emp Id":["101","102","103","104","105","106"],
       "Salary":[40000,50000,60000,70000,80000,90000]}
```

```python
df1=pd.DataFrame(data1)
df2=pd.DataFrame(data2)
```

```python
df1
```

```python
df2
```

```python
print(pd.merge(df1,df2,on="Emp Id"))   # Merge the DataFrame
```

# if the dataframe have not same value(means Emp Id are different in some cases)

```python
data1={"Emp Id":["101","108","103","104","105","106"],
       "Name":["khlaid","jameel","Umar","Javid","Haseeb","ashar"],
       "Age":[20,30,40,50,60,45]}
```

```python
data2={"Emp Id":["101","102","103","104","109","106"],
       "Salary":[40000,50000,60000,70000,80000,90000]}
```

```python
df1=pd.DataFrame(data1)
df2=pd.DataFrame(data2)
```

```python
print(pd.merge(df1,df2,on="Emp Id"))   # whose emp id is match return the result
```

```python
print(pd.merge(df1,df2,how="inner"))    # inner merge(whose emp id is match then return the result)
```

```python
print(pd.merge(df1,df2,how="left"))  # left merge(left table complete & right table whose value is match)
```

```python
print(pd.merge(df1,df2,how="right"))  # right merge(right table complete & left table whose value is match)
```

# Pivoting and Melting DataFrame

```python
import pandas as pd
```

```python
data={"Keys":["k1","k2","k1","k2"],
      "Names":["khlaid","Ali","umar","junaid"],
      "Houses":["red","blue","green","red"],
      "Grades":["A","B","C","D"]}
```

```python
df=pd.DataFrame(data)
```

```python
data
```

```python
df
```

```python
print(df.pivot(index="Keys",columns="Names",values="Houses")) # Poivt one colum(Names become column name)
```

```python
print(df.pivot(index="Keys", columns="Names", values=["Houses", "Grades"])) #povit two column
```

# Melting In Pandas

```python
data={
      "Names":["khlaid","Ali","umar","junaid"],
      "Houses":["red","blue","green","red"],
      "Grades":["A","B","C","D"]}
```

```python
df=pd.DataFrame(data)
```

```python
df
```

```python
print(pd.melt(df,id_vars=["Names"],value_vars=["Houses"]))  # melt split the column Houses into two column(one is variable and second is value)
```
